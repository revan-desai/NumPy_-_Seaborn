```python
import seaborn as sns

import matplotlib.pyplot as plt

import pandas as pd

v=[1,5,4,2,3]

v1=[1,2,3,4,5]

data=pd.DataFrame({'var':v,'var1':v1})

sns.lineplot(x='var',y='var1',data=data)



```

```python
import seaborn as sns

import matplotlib.pyplot as plt

data=sns.load_dataset('iris')

sns.lineplot(x='petal_length',y='petal_width',data=data,hue='sepal_width')  # try palette='Accent'

plt.show()
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

data=sns.load_dataset('iris')

sns.lineplot(x='petal_length',y='petal_width',data=data,hue='sepal_width',palette='Accent')  # try palette='Accent' ,'bright','muted'

plt.show()
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

data=sns.load_dataset('iris')

sns.set_style('darkgrid')

sns.lineplot(x='petal_length',y='petal_width',data=data,hue='species',alpha=0)

```
