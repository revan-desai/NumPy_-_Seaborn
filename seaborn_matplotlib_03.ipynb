## Creating Different Types of Plots

## relplot()-- relational plot rel plot

## remember -- the y axis value has to be a  categorical and not continous numerical

```python
import seaborn as sns

import matplotlib.pyplot as plt

datairis=sns.load_dataset('iris')

sns.relplot(x='sepal_width',y='species',data=datairis)

plt.show()
```

## scatter plot()

```python
import seaborn as sns

import matplotlib.pyplot as plt

datairis=sns.load_dataset('iris')

sns.scatterplot(x='sepal_length', y='sepal_width',data=datairis)

plt.show()

```

## For certain datasets, you may want to consider changes as a function of time in one variable, or as a similarly continuous variable.

e.g stock charts

```python
import seaborn as sns

import matplotlib.pyplot as plt

datairis=sns.load_dataset('iris')

sns.lineplot(x='sepal_length', y='species',data=datairis)   # you can take any values on x and y

plt.show()
```
