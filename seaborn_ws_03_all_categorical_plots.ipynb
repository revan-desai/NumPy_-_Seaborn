```python
import seaborn as sns

import matplotlib.pyplot as plt

datatip=sns.load_dataset('tips')

sns.catplot(x='day',y='tip',data=datatip,palette='bright')
```

```python
sns.swarmplot(x='day',y='tip',data=datatip,hue='sex',palette='bright')
```

## box plot

```python
import seaborn as sns

import matplotlib.pyplot as plt

datatip=sns.load_dataset('tips')

sns.boxplot(x='day',y='total_bill',hue='sex',data=datatip,palette='bright')
```

## violin plot

```python
sns.violinplot(x='day',y='tip',hue='sex',split=True,data=datatip)
```
