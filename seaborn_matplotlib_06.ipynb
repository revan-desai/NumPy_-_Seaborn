## Violinplot
It is similar to the boxplot except that it provides a higher, more advanced visualization and uses the kernel density estimation to give a better description about the data distribution. It is created using the violinplot() method.

```python
import seaborn as sns

import matplotlib.pyplot as plt

datairis=sns.load_dataset('iris')

sns.violinplot(x='species',y='sepal_length',data=datairis)      # no complusion of having categorical value at x or y axis

plt.show()


```

## stripplot()--> it basically creates a scatter plot based on the category. It is created using the stripplot() method.

```python
import seaborn as sns

import matplotlib.pyplot as plt

datairis=sns.load_dataset('iris')

sns.stripplot(x='species',y='petal_width',data=datairis,palette='bright')   #palette ='muted','dark'   etc

plt.show()
```

## swarmplot()--> Swarmplot is very similar to the stripplot except the fact that the points are adjusted so that they do not overlap

```python
import seaborn as sns

import matplotlib.pyplot as plt

sns.swarmplot(x='species',y='petal_width',data=datairis,palette='bright')

plt.show()
```

## cat plot()

```python
import seaborn as sns

import matplotlib.pyplot as plt

sns.catplot(x='species',y='petal_width',data=datairis,palette='bright')

plt.show()
```

## displot()

```python
import seaborn as sns
import matplotlib.pyplot as plt

# loading dataset
data = sns.load_dataset("iris")

sns.displot(data['sepal_width'])
plt.show()
```
