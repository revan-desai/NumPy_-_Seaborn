## Multiple plots with Seaborn

## using add_axes()

```python
import seaborn as sns

import matplotlib.pyplot as plt

datairis=sns.load_dataset('iris')

def graph():
 sns.lineplot(x="sepal_length", y="sepal_width", data=datairis)

fig1=plt.figure(figsize=(6,6))   #(width,height) in inches

fig1.add_axes([0.1,0.1,1,1])  #[x0,y0,width,height]

graph()

fig1.add_axes([0.5,0.7,0.2,0.2])     # inner plot width and height are 20% of outer one. if you increase second value 0.7 to 0.9 inner plot will move upward
                                     # and if you increase first value 0.5 to 0.8 then inner plot will move rightward.
graph()

plt.show()


```

## using subplot method()

```python
import seaborn as sns

import matplotlib.pyplot as  plt

datairis=sns.load_dataset('iris')

def graph():
  sns.lineplot(x='sepal_length',y='sepal_width',data=datairis)

plt.subplot(121)

graph()

plt.subplot(122)   # other numbers dont work 123 124

graph()

plt.show()

```

## Using FacetGrid() method

```python
import seaborn as sns

import matplotlib.pyplot as plt

datairis=sns.load_dataset('iris')

plot123=sns.FacetGrid(datairis,col='species')   # The column variables should be categorical....and not 'continous' numerical values like 1,2,3,4,5.......
                                                # e.g. (male,female).....category like(different country code ..+91 +1 +62 this is also kind of category)

plot123.map(plt.plot,'sepal_width')

plt.show()
```
