## sns.lineplot(x='',y='',data=name of data set)

```python
import seaborn as sns

import matplotlib.pyplot as plt

datairis=sns.load_dataset("iris")

sns.lineplot(x='sepal_length',y='sepal_width',data=datairis)
```

## Setting the xlim and ylim

```python
import seaborn as sns

import matplotlib.pyplot as plt

datairis=sns.load_dataset('iris')

sns.lineplot(x='sepal_length',y='sepal_width',data=datairis)

plt.xlim(5)         # values on x axis will start from 5

plt.show()

```

## Customizing Seaborn Plots

There are five themes available in Seaborn.

1. darkgrid
2. whitegrid
3. dark
4. white
5. ticks

```python
import seaborn as sns

data=sns.load_dataset('iris')

sns.set_style('darkgrid')          # other 4 arguments .....whitegrid,dark,white,ticks

sns.lineplot(x='sepal_length',y='sepal_width',data=datairis)
```
