```python
from pandas.core.groupby.generic import DataFrameGroupBy

import seaborn as sns

import matplotlib.pyplot as plt

tdata=sns.load_dataset('titanic')

sns.displot(tdata['age'])

#OR

# sns.displot(x='age',data=tdata)     # not providing y axis value
```

```python
sns.kdeplot(tdata['age'])    # means the probability of finding a person with age range 20-40 is high because density is high
```

```python

import seaborn as sns

import matplotlib.pyplot as plt

datairis=sns.load_dataset('iris')

sns.kdeplot(datairis['petal_width'])

plt.show()
```

```python

import seaborn as sns

import matplotlib.pyplot as plt

datairis=sns.load_dataset('iris')

sns.kdeplot(x='sepal_length',data=datairis)     # f you dont provide y axis value then it will show density

plt.show()
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

tdata=sns.load_dataset('titanic')

sns.kdeplot(tdata[tdata['survived']==1]['age'],color='green')

sns.kdeplot(tdata[tdata['survived']==0]['age'],color='red')

plt.show()
```
